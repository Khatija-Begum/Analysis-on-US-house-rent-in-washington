```python
#creating a new column to save the other data
df2=df[df['sqft_living']<=6000]
```

```python
df2.head()
```

```python
#creating a class intervals to create a proper bargraph

sqft_bins=[0,1000,2000,3000,4000,5000,6000]  #in bins we write the total intervals or the numbers
sqft_labels=['0-1000','1000-2000','2000-3000','3000-4000','4000-5000','5000-6000']     #class interval in which form we need it

df2['SqftGroup'] = pd.cut(df['sqft_living'],bins=sqft_bins,labels=sqft_labels)


```

```python
df2.sample(5)
```

```python
sns.barplot(x=df2['SqftGroup'],y=df2['price'])
```

```python
#INSIGHT
# Here we can that ,as the sqft increases prices also increases.
# Mostly the houses were having the sqft between 2000-4000.
```
